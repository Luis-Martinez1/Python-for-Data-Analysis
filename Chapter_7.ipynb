# lambda expression



```python
def sum_two_numbers(x, y):
    return x + y
```

```python
sum_two_numbers(4, 5)
```

```python
sum_2_num = lambda x, y: x + y
```

```python
sum_2_num(5, 6)
```

```python
pass_fail = lambda x: 'Pass' if x>=60 else 'Fail'
```

```python
pass_fail(90)
```

```python
pass_fail(59)
```

```python
def grade_to_letter(grade):
    if grade >= 90:
        return 'A'
    elif grade >= 80:
        return 'B'
    elif grade >= 70:
        return 'C'
    elif grade >= 60:
        return 'D'
    else:
        return 'F'
```

```python
grade_to_letter(67)
```

```python
g_2_l = lambda g: 'A' if g>=90 else ('B' if g>=80 else ('C' if g>=70 else ('D' if g>=60 else 'F')))
```

```python
g_2_l(67)
```
