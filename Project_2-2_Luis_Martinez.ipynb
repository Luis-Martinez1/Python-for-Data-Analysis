```python
# 8. Multiply the Total Volume and AveragePrice columns, and store the result in a new
# column called EstimatedRevenue. Then, display the first five rows of this data to
# confirm that the column was added and has the correct values.



# Multiply 'Total Volume' and 'AveragePrice' columns to create 'EstimatedRevenue'
df['EstimatedRevenue'] = df['Total Volume'] * df['AveragePrice']

# Display the first five rows with the new column
df.head()



```

```python
# 9. Create a DataFrame that’s grouped by region and type and that includes the average
# price for the grouped columns. Then, reset the index and display the first five rows.

# Create a DataFrame grouped by 'region' and 'type' with the average price
grouped_df = df.groupby(['region', 'type'])['AveragePrice'].mean().reset_index()

# Display the first five rows of the grouped DataFrame
grouped_df.head()
```

```python
# 10. Create a bar plot that shows the mean, median, and standard deviation of the Total Volume column by year.

#import matplotlib.pyplot as plt

# Convert the 'Date' column to datetime format to extract the year
# df['Date'] = pd.to_datetime(df['Date'])
# df['Year'] = df['Date'].dt.year

# Group the data by 'Year' and calculate mean, median, and standard deviation of 'Total Volume'
summary_by_year = df.groupby('Year')['Total Volume'].agg(['mean', 'median', 'std'])

# Plot the bar chart
summary_by_year.plot()#(kind='bar', yerr='std', capsize=5, figsize=(10, 6))
# plt.title('Mean, Median, and Standard Deviation of Total Volume by Year')
# plt.xlabel('Year')
# plt.ylabel('Total Volume')
# plt.legend(['Mean', 'Median', 'Std Dev'])
# plt.show()
```

```python
# 1. How many unique regions are there?
# Count the unique regions
num_unique_regions = df['region'].nunique()

# Display the result
print("Number of unique regions:", num_unique_regions)
```

```python
# 2. What is the average price for each type of avocado (organic and conventional)? Be
# sure to include just the type and AveragePrice columns in the results.


# Group the data by 'type' and calculate the mean of 'AveragePrice'
average_price_by_type = df.groupby('type')['AveragePrice'].mean().reset_index()

# Display the result
print("Average price for each type of avocado:")
print(average_price_by_type[['type', 'AveragePrice']])

```

```python
# 3. Which region has the lowest average price for organic avocados? Hint: Create wide
# data from the grouped data that you created in task 8.


# Group the data by 'region' and 'type' and calculate the mean of 'AveragePrice'
grouped_by_region_type = df.groupby(['region', 'type'])['AveragePrice'].mean().reset_index()

# Filter the DataFrame to include only organic avocados
organic_prices = grouped_by_region_type[grouped_by_region_type['type'] == 'organic']

# Find the region with the lowest average price for organic avocados
lowest_avg_price_region = organic_prices.loc[organic_prices['AveragePrice'].idxmin()]

# Display the result
print("Region with the lowest average price for organic avocados:")
print(lowest_avg_price_region[['region', 'AveragePrice']])

```

```python
# 4. Have the Total Bags sold per year of each type of avocado become more or less
# consistent over time?


# Convert the 'Date' column to datetime format to extract the year
# df['Date'] = pd.to_datetime(df['Date'])
# df['Year'] = df['Date'].dt.year

# Group the data by 'Year' and 'type' and calculate the coefficient of variation for 'Total Bags'
cv_by_year_type = df.groupby(['Year', 'type'])['Total Bags'].std() / df.groupby(['Year', 'type'])['Total Bags'].mean()

# Display the result
print("Coefficient of Variation for Total Bags sold per year of each type of avocado:")
print(cv_by_year_type)

```
