```python
# imports
import pandas as pd
```

```python
# 1. Read the data from the CSV file into a DataFrame and display the first five rows.
# 1. Read the data from the CSV file into a DataFrame.

# exams file is in the local working directory
file_path = 'exams.csv'

# Use pandas to read the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify the import
df.head()
```

```python
# 2. Display the basic information for the DataFrame and its columns using the info() method.
df.info()
```

```python
# 3. Display statistical information for the math score, reading score, and writing score columns using 
# the describe() method.

df.describe().T
```

```python
# 4. Group the data by the race/ethnicity column and display the mean scores.
# Group the data by 'race/ethnicity' and calculate the mean scores
mean_scores_by_ethnicity = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()

# Display the result
print("Mean scores by race/ethnicity:")
print(mean_scores_by_ethnicity)
```

```python
# 5. Display a single column as a DataFrame with bracket notation.
# Display the 'math score' column
math_scores_df = df[['math score']]

math_scores_df
```

```python
# 6. Display a single column as a Series with bracket notation.

math_scores_series = df['math score']

math_scores_series
```

```python
# 7. Display a single column as a Series with dot notation.
gender_series_dot = df.gender

gender_series_dot
```

```python
# 8. Display only rows for females with a math score greater than or equal to 90.

# gender female and math score above 90
female_high_math_scores = df[(df['gender'] == 'female') & (df['math score'] >= 90)]


female_high_math_scores
```

```python
# 1. Does taking a test preparation course improve average scores?

df['test preparation course'].value_counts()

test_prep = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
test_prep
```

```python
# 2. Which gender is better on average at math?
# Group the data by 'gender' and calculate the mean math score
average_math_score_by_gender = df.groupby('gender')[['math score']].mean()

average_math_score_by_gender
```
