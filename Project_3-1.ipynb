# Plot the Ramen Data with Pandas

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Read the data from the CSV file into a DataFrame.

# ramen-ratings file is in the local working directory
file_path = 'ramen-ratings.csv'

# Use pandas to read the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify the import
df.head()
```

```python
# 2. Create a histogram for the Stars column using the number of bins necessary to
#    display the ratings for each quarter of a star.

# Define the number of bins based on the range of stars
num_bins = int((df['Stars'].max() - df['Stars'].min()) * 4) 

# Create the histogram
plt.hist(df['Stars'], bins=num_bins)

# Add labels and title
plt.xlabel('Stars')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')

# Show the histogram
plt.show()

```

```python
# 3. Create a KDE plot for the Stars column.

# Create the KDE plot
sns.kdeplot(df['Stars'], fill=True)

# Add labels and title
plt.xlabel('Stars')
plt.ylabel('Density')
plt.title('Kernel Density Estimation (KDE) Plot for Ratings')

# Show the plot
plt.show()
```

```python
# 4. Create a bar plot that shows the average number of stars by ramen style. Hint: Youll
# need to group the data before you plot it.

# Grouping by 'Style' and calculating the mean of 'Stars' for each group, then plotting a bar plot
df.groupby('Style')['Stars'].mean().plot.bar()

# Adding labels and title
plt.xlabel('Ramen Style')
plt.ylabel('Average Rating')
plt.title('Average Rating for Each Ramen Style')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Showing the plot
plt.show()
```

```python
# 5. Create a bar plot that shows the total number of reviews for each ramen type.

# Grouping by 'Style' and counting the occurrences of 'Stars' for each group, then plotting a bar plot
df.groupby('Style')[['Stars']].count().plot.bar()

# Adding labels and title
plt.xlabel('Ramen Style')
plt.ylabel('Number of Reviews')
plt.title('Total Number of Reviews for Each Ramen Style')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Showing the plot
plt.show()
```

```python
# 6. Create a bar plot that shows the highest score given to each ramen type.

# Grouping by 'Style' and finding the maximum value of 'Stars' for each group, then plotting a bar plot
df.groupby('Style')[['Stars']].max().plot.bar()

# Adding labels and title
plt.xlabel('Ramen Style')
plt.ylabel('Highest Rating')
plt.title('Highest Rating Given to Each Ramen Style')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Showing the plot
plt.show()
```

```python
# 1. What is the most common style of ramen?

# Count the occurrences of each ramen style
common_style = df['Style'].value_counts().idxmax()

print("The most common style of ramen is:", common_style)
```
