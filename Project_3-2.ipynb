# Plot the Avodaco Data with Pandas

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Read the data from the .csv file into a DataFrame and display the first five rows.

df = pd.read_csv('avocado.csv')
df.head()
```

```python
# 2. Create a new DataFrame that contains the total of the Small Bags, Large Bags, and
# XLarge Bags columns grouped by type, and then display the DataFrame.

# Grouping by 'type' and summing the 'Small Bags', 'Large Bags', and 'XLarge Bags' columns
bags_totals = df.groupby('type')[['Small Bags', 'Large Bags', 'XLarge Bags']].sum()
bags_totals
```

```python
# 3. Use the grouped data to create a bar plot that shows the number of small, large, and
# extra-large bags for both types of avocado.
bags_totals = df.groupby('type')[['Small Bags', 'Large Bags', 'XLarge Bags']].sum().plot.bar()
```

```python
# 4. Use the original data to create a scatter plot for the Total Volume and AveragePrice columns.

# Creating the scatter plot
plt.scatter(df['Total Volume'], df['AveragePrice'], alpha=0.5, color='blue')

# Adding labels and title
plt.xlabel('Total Volume')
plt.ylabel('Average Price')
plt.title('Scatter Plot: Total Volume vs Average Price')

# Showing the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 1. In what year were the most bags of avocados sold?

# Grouping by 'Year' and summing the 'Total Bags' column
bags_by_year = df.groupby('year')['Total Bags'].sum()

# Finding the year with the highest total bags sold
year_most_bags_sold = bags_by_year.idxmax()

print("The year with the most bags of avocados sold was:", year_most_bags_sold)
```

```python
# OR use visual
df.groupby('year')['Total Bags'].sum().plot.bar()
```
