```python
# imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Read the data from the CSV file into a DataFrame and display the first five rows.

# avocado file is in the local working directory
file_path = 'avocado.csv'

# Use pandas to read the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify the import
df.head()
```

```python
# 2. Create the following plot with a Seaborn specific method.


sns.scatterplot(x='AveragePrice', y='Total Volume', data=df)

# Set plot labels and title
plt.xlabel('Average Price')
plt.ylabel('Total Volume')
plt.title('Scatter Plot: Total Volume vs Average Price')

# Show the plot
plt.show()
```

```python
# 3. Create the same plot with a generic method.

sns.relplot(x='AveragePrice', y='Total Volume', data=df, kind='scatter')

# Set plot labels and title
plt.xlabel('Average Price')
plt.ylabel('Total Volume')
plt.title('Scatter Plot: Total Volume vs Average Price')

# Show the plot
plt.show()
```

```python
# 4. Create the same plot but with the hue parameter set to year and the dots for the total
#   volume ranging in size from 10 to 100.


sns.scatterplot(x='AveragePrice', y='Total Volume', hue='year', size='Total Volume', sizes=(10, 100), data=df)

# Set plot labels and title
plt.xlabel('Average Price')
plt.ylabel('Total Volume')
plt.title('Scatter Plot: Total Volume vs Average Price with Year Hue')

# Show the plot
plt.show()
```

```python
# 5. Create a plot that looks like the one that follows.

regions_to_plot = ["TotalUS", "West", "WestTexNewMexico"]
filtered_df = df[df['region'].isin(regions_to_plot)]

# Set a custom palette with red, orange, and blue colors
custom_palette = {'TotalUS': 'red', 'West': 'orange', 'WestTexNewMexico': 'blue'}

# Create a figure-level bar chart
sns.catplot(x='year', y='Total Volume', hue='region', data=filtered_df, kind='bar', palette=custom_palette, ci=90)

# Set plot labels and title
plt.xlabel('Year')
plt.ylabel('Total Volume')
plt.title('Stacked Bar Chart: Total Volume by Year and Region')

# Show the plot
plt.show()
```
