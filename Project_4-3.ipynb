```python
# imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Read the data from the CSV file into a DataFrame and display the first five rows.

# exams file is in the local working directory
file_path = 'exams.csv'

# Use pandas to read the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify the import
df.head()
```

```python
# 2. Use the Seaborn catplot() method to create a plot like the one below. To make that easy
# to do, you can use a count plot.

sns.catplot(x='parental level of education', kind='count', data=df)

# Set plot labels and title
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')
plt.title('Count Plot of Parental Level of Education')

# Show the plot
plt.show()
```

```python
# 3. Rotate the x labels for the above plot to make them readable.

sns.catplot(x='parental level of education', kind='count', data=df)

# Set plot labels and title
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')
plt.title('Count Plot of Parental Level of Education')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()

```

```python
# 4. Create a scatter plot with a specific method that compares the writing score with the
# reading score. Use a different color to show which students took a test prep course.

sns.scatterplot(x='reading score', y='writing score', hue='test preparation course', data=df)

# Set plot labels and title
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.title('Scatter Plot: Writing Score vs Reading Score with Test Prep Course')

# Show the plot
plt.show()

```

```python
# 5. Create the same plot with a general method.
sns.relplot(x='reading score', y='writing score', hue='test preparation course', kind='scatter', data=df)

# Set plot labels and title
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.title('Scatter Plot: Writing Score vs Reading Score with Test Prep Course')

# Show the plot
plt.show()
```

```python
# 6. Adjust the size of the scatter plot you created with the general method so its more like
# the size of the scatter plot with the specific method.


sns.relplot(x='reading score', y='writing score', hue='test preparation course', kind='scatter', data=df, height=4)

# Set plot labels and title
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.title('Scatter Plot: Writing Score vs Reading Score with Test Prep Course')

# Show the plot
plt.show()
```
